```python
#!/usr/bin/python
# -*- coding: utf-8 -*-

import os
import csv
```

```python
def getDistance(file):
    f = open(file,'r')
    average_distance = 0
    count = 0
    for x in f:
        dist_at_frame = float(x.split()[1])
        if dist_at_frame < 5:
            average_distance += dist_at_frame
            count += 1
    return (average_distance)/count

def getAllDistances(file):
    f = open(file,'r')
    distances = []
    for x in f:
        dist_at_frame = float(x.split()[1])
        if dist_at_frame < 5:
            distances.append(dist_at_frame)
    return distances

```

```python
directory = '../PDB/'

count = 0
filedNames = ["acceptor_chain_id","acceptor_type" , "acceptor_id" ,"donor_chain_id","donor_type" ,"donor_id","distance"]

f = open('salt_bridges_data.csv', 'w') 
writer = csv.DictWriter(f,fieldnames = filedNames) 
writer.writeheader()
for foldername in os.scandir(directory):
    if foldername.name != ".DS_Store": 
        for folders in os.scandir(foldername):

            if folders.is_dir() and folders.name == "salt_bridges":
                count += 1 
                
                for files in os.scandir(folders):
                    if files.name[-1:-4:-1] == "tad":
                        dist = getDistance(files)
                        #print(files.name)
                        l = files.name.split("_")

                        acceptor_chain_id = l[1][5]
                        acceptor_type = l[0][7:10]
                        acceptor_id = l[0][10::]

                        donor_chain_id = l[2][5]
                        donor_type = l[1][7:10]
                        donor_id = l[1][10::]


                        print(acceptor_chain_id,acceptor_id,acceptor_type , donor_chain_id,donor_id,donor_type)

                        
                        writer.writerow({"acceptor_chain_id" : acceptor_chain_id ,"acceptor_type" : acceptor_type,"acceptor_id" : acceptor_id,"donor_chain_id" : donor_chain_id,"donor_type":donor_type, "donor_id": donor_id,"distance":dist}) 
```
